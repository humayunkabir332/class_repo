```python
email = input("Enter email: ").strip()


if not email:
    print("Invalid Email: empty input")
else:
    at_count = 0
    for ch in email:
        if ch == "@":
            at_count += 1

    
    if at_count != 1:
        print("Invalid Email: must contain exactly one '@'")
    else:
       
        at_index = email.find("@")

      
        if at_index == 0 or at_index == len(email) - 1:
            print("Invalid Email: nothing before or after '@'")
        else:
            
            if email.endswith(".com") or email.endswith(".org"):
                print("Valid Email")
            else:
                print("Invalid Email: must end with .com or .org")
```
